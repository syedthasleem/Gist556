## Boxplot
- also known as whgisker plot
- parameters
- min(1-5%)
- first quirtile(25%)
- second quirtile(25%)
- max quirtile

```python
data=np.random.normal(100,20,200)
data
```

```python
plt.boxplot(data)
plt.show()
```

```python
data1=np.random.randint(50,100,10)
print(data1)
data2=np.random.randint(40,90,10)
data3=np.random.randint(34,109,10)
dx=[data1,data2,data3]
dy=['a','b','c']
plt.boxplot(dx,dy)
plt.plot()
plt.savefig('boxplot.png')
```

## Pie chart
- division of each peace of data into separate and visualize

```python
branches_c=[500,450,800,250,300]
branches=['ECE','CIVIL','CSE','MECH','EEE']
colors=['c','r','m','b','y','g']
plt.pie(branches_c,labels=branches,startangle=75,shadow=True,explode=(0.1,0,0.2,0,0),autopct='%1.1f%%')
plt.show()
```
